```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import plotly.express as px
grafico =  px.histogram (df,  x = "Nota", nbins=20, ) 
grafico.update_layout(width=500,height=500,title_text='Distribuição de Notas') 
grafico.show()
```

Mostrando a distribuição de notas, podemos observar que a grande maioria dos filmes tem nota 4 de 5

```python
plt.hist(df.Nota, bins='sturges', rwidth=1)
plt.title('Distribuição de Notas')
plt.show()
```

```python
sns.displot(df['Nota'], bins=20, color="brown", kde=True);
```

A maior nota é 5, o que não é surpresa pois muitos filmes são muito bons, mas a menor nota não ser zero significa que mesmo o pior dos filmes, tem qualidades que fazem ter pelo menos nota 0.5

```python
plt.hist(df.Nota.min(), bins=5, rwidth=1)
plt.hist(df.Nota.max(), bins=5, rwidth=1)
plt.title('Menor e Maior Nota registrada')
plt.show()
```

Por meio do box plot podemos observar que a maioria das notas estão divididas entre 3 e 4.

```python
fig = px.box(df, y="Nota", points="all")
fig.show()
```

Podemos ver alguns outliers no 1 e na região do 0.3, o que significa que poucos são os filmes com notas muito baixas.
