```python
import pandas as pd

%matplotlib inline
```

```python
filmes_csv = 'Dataset/movies.csv'
ratings_csv = 'Dataset/ratings.csv'



filmes_df = pd.read_csv(filmes_csv)
ratings_df = pd.read_csv(ratings_csv)
```

```python
filmes_df.head()
```

```python
filmes_df.info()
```

## Vamos renomear o nome das colunas para ficar mais entendivel

```python
filmes_df.rename(columns={'e (2016)': 'Nome', 'Crime|Drama': 'Genero'}).head()
```

## Juntando o dataset de filmes com dataset de notas 


```python
df = filmes_df.merge(ratings_df,on=["movieId"])
df.head()
```

## Renomeando as colunas do novo dataset conjunto

```python
df = df.rename(columns={'movieId':'Filme ID','title': 'Nome', 'genres': 'Genero', 'userId': 'ID Usu√°rio', 'rating': 'Nota'})

df
```

## Procurando por linhas com valores nulos

```python
df.isnull().sum()
```

## Removendo linhas com valores nulos
Por meio de analize observamos que a coluna as notas possue um valor nulo, portanto, iremos remove-lo.


```python
df.dropna(subset=['Nota'], inplace=True)
```

```python
df['Genero'].to_csv('Dataset/genero.csv', index=False)
```
