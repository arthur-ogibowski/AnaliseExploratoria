## Traduzir e listar G√™neros
Na func, recebemos a coluna de genero, e a op√ß√£o de listar ou traduzir, se escolhido listar, as linhas ser√£o transformadas em arrays e iteradas sobre, se ja houver o genero na lista, n√£o ser√° adicionado, assim podemos ver todos os generos individuais. Se escolhido traduzir, iteramos sobre as linhas e fazemos substitui√ß√£o das substrings.

```python
def limpar_genero(coluna, option):

  if option is "listar":
    result_string = coluna.str.lower().str.split("|")
    lista_generos = []

    for generos in result_string:
      for genero in generos:
        if genero not in lista_generos:
          lista_generos.append(genero)

    print(lista_generos)

  elif option is "traduzir":

    for linha in coluna:
      print(linha.replace('Adventure', 'Aventura').replace('Animation', 'Anima√ß√£o').replace('Children', 'Infantil').replace('Comedy', 'Comedia').replace('Fantasy', 'Fantasia').replace('Action', 'A√ß√£o').replace('Thriller', 'Suspense').replace('Horror', 'Terror').replace('Mystery', 'Mist√©rio').replace('Sci-Fi', 'Fic√ß√£o Cient√≠fica').replace('Documentary', 'Document√°rio').replace('War', 'Guerra').replace('Western', 'Velho-Oeste').replace('Film-noir', 'Detetive').replace('(no genres listed)', 'Sem g√™nero'))
  
```

```python
limpar_genero(filmes_df['Genero'], "listar")
```

```python
limpar_genero(filmes_df['Genero'], "traduzir")
```
